```python
# Dependencies
import pandas as pd
import numpy as np
from functools import reduce
import matplotlib.pyplot as plt
```

```python
#Get html for name
url = 'https://en.wikipedia.org/wiki/The_Simpsons'

#Get results
results=pd.read_html(url)
```

```python
#Set series overview table as df
overview=results[1]
```

```python
#Read columns
overview.columns
```

```python
#Merge multi-index
overview.columns = [' '.join(col).strip() for col in overview.columns.values]
```

```python
#Read collapsed columns
overview.columns
```

```python
#Rename columns
overview=overview.rename(columns = {'Season Season Season':'Season',
                                    'Season Season Season.1': 'Years',
                                   'No. ofepisodes No. ofepisodes No. ofepisodes': 'No. of Episodes',
                                   'Originally aired Season premiere Season premiere': 'Original Premiere Date',
                                   'Originally aired Season finale Season finale': 'Original Finale Date',
                                    'Originally aired Time Slot (ET) Time Slot (ET)': 'Original Time Slot(ET)',
                                    'Viewership Avg. viewers(in millions) Avg. viewers(in millions)': 'Ave. Viewers in mil',
                                   'Viewership Most watched episode Viewers(millions)':'Highest Viewers in mil',
                                    'Viewership Most watched episode Episode Title':'Most Watched Title'})
```

```python
overview
```

```python
#Remove extra string from average viewers
overview['Ave. Viewers in mil']=overview['Ave. Viewers in mil'].str[:4]

#Remove second year from season year
overview['Years']=overview['Years'].str[:4]

#Remove [ from 26
overview['Ave. Viewers in mil'][26]=overview['Ave. Viewers in mil'][26][:-1]
```

```python
#Drop index 31
overview=overview.drop(31)
```

```python
#Set columns as integers
overview['Season'] = overview['Season'].astype(int)
overview['Years'] = overview['Years'].astype(int)
overview['No. of Episodes'] = overview['No. of Episodes'].astype(int)
overview['Ave. Viewers in mil'] = overview['Ave. Viewers in mil'].astype(float)
overview['Highest Viewers in mil'] = overview['Highest Viewers in mil'].astype(float)
```

```python
#Add one to year
overview['Years'] = overview['Years'] +1
```

```python
overview
```

```python
All_Seasons=pd.DataFrame({'Season_Number': overview['Season'],
                        'Season_Year_Aired ': overview['Years']})
```

```python
All_Seasons
```

```python
#Write data frame to csv
All_Seasons.to_csv("./Season_Year.csv", sep=',',index=False)
```
